# EDA-Based Assumption
Based on the EDA results, **sleep hours** show a negative relationship with **stress level**.
The study aims to test whether sleep hours significantly affect stress level by using Pearson correlation analysis and hypothesis testing.


# Load Data


```python
df = pd.read_csv('../../00-dataSource/clean_data_v1.csv')
df.head()
```

In this case, we test the relationship between **sleep_hours** and **stress_level**.

Hypotheses:
- **H0**: There is no linear relationship between sleep hours and stress level.
- **H1**: There is a linear relationship between sleep hours and stress level.

Decision rule:
- If `p-value < 0.05`, reject **H0**.
- If `p-value >= 0.05`, fail to reject **H0**.


```python
pearson_results = pd.DataFrame([
    pearson_test(df, 'sleep_hours', 'stress_level'),
])

pearson_results
```

## Conclusion
Based on the test result, the relationship between **sleep_hours** and **stress_level** is **statistically significant** because the p-value is less than `0.05`. The correlation is **negative**, which means that as sleep hours increase, stress level tends to decrease. Therefore, we **reject H0**. This means the negative relationship observed in EDA is **supported by statistical testing**.
