# EDA-Based Assumption
Based on the EDA results, **device usage** shows a weak positive relationship with **productivity score**.
The study aims to test whether device usage significantly affects productivity by using Pearson correlation analysis and hypothesis testing.


# Load Data


```python
df = pd.read_csv('../../00-dataSource/clean_data_v1.csv')
df.head()
```

In this case, we test the relationship between **device_hours_per_day** and **productivity_score**.

Hypotheses:
- **H0**: There is no linear relationship between device hours per day and productivity score.
- **H1**: There is a linear relationship between device hours per day and productivity score.

Decision rule:
- If `p-value < 0.05`, reject **H0**.
- If `p-value >= 0.05`, fail to reject **H0**.


```python
pearson_results = pd.DataFrame([
    pearson_test(df, 'device_hours_per_day', 'productivity_score'),
])

pearson_results
```

## Conclusion
Based on the test result, the relationship between **device_hours_per_day** and **productivity_score** is **statistically significant** because the p-value is less than `0.05`. However, the correlation is **very weak positive**, which means the relationship exists statistically but its practical effect is small. Therefore, we **reject H0**. This means the EDA pattern is **supported statistically**, but the actual strength of the relationship is weak.
