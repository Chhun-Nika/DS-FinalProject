# EDA-Based Assumption
From the EDA step, we want to check whether **gender** shows a meaningful difference in **productivity_score**.
This is the relationship suggested by exploration, and hypothesis testing is used to verify whether that observed pattern is statistically supported.


# Load Data


```python
df = pd.read_csv('../../00-dataSource/clean_data_v1.csv')
df.head()
```

In this case, we test the relationship between **gender** and **productivity_score**.

Hypotheses:
- **H0**: The mean productivity score of the two gender groups is equal.
- **H1**: The mean productivity score of the two gender groups is different.

Decision rule:
- If `p-value < 0.05`, reject **H0**.
- If `p-value >= 0.05`, fail to reject **H0**.


```python
ttest_results = pd.DataFrame([
    independent_t_test(df, 'gender', 'productivity_score', 'female', 'male'),
])

ttest_results
```

## Conclusion
Based on the test result, the difference in **productivity_score** between female and male groups is **not statistically significant** because the p-value is greater than `0.05`. Therefore, we **fail to reject H0**. This means the EDA pattern is **not supported strongly enough by statistical testing**, and gender does not show a meaningful difference in productivity score in this dataset.
