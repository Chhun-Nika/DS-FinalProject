# EDA-Based Assumption
From the EDA step, we want to check whether **gender** shows a meaningful difference in **stress_level**.
This is the relationship suggested by exploration, and hypothesis testing is used to verify whether that observed pattern is statistically supported.


# Load Data


```python
df = pd.read_csv('../../00-dataSource/clean_data_v1.csv')
df.head()
```

In this case, we test the relationship between **gender** and **stress_level**.

Hypotheses:
- **H0**: The mean stress level of the two gender groups is equal.
- **H1**: The mean stress level of the two gender groups is different.

Decision rule:
- If `p-value < 0.05`, reject **H0**.
- If `p-value >= 0.05`, fail to reject **H0**.


```python
ttest_results = pd.DataFrame([
    independent_t_test(df, 'gender', 'stress_level', 'female', 'male'),
])

ttest_results
```

## Conclusion
Based on the t-test results, there is no statistically significant difference in stress levels between male and female groups in this dataset.

This suggests that gender is not a strong factor influencing stress level in this study.
