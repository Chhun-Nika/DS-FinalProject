# load dataset

## 2. Explorary Data Analysis

### 2.1 Distribution Analysis
The purpose of this step is to understand how each numerical feature is distributed across the dataset.

```python
cols = [
    'device_hours_per_day',
    'notifications_per_day',
    'social_media_mins',
    'study_mins',
    'sleep_hours',
    'stress_level',
    'digital_dependence_score'
]

for col in cols:
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

### Plot Interpretation

#### Device Hours Per Day
The distribution of **device_hours_per_day** is slightly right-skewed. Most users spend around **5 to 8 hours per day** on their devices, while a smaller number of users have much higher screen time.

#### Notifications Per Day
The **notifications_per_day** feature is strongly right-skewed. Most users receive a moderate number of notifications, but a few users receive very high notification counts, creating a long tail.

#### Social Media Minutes
The distribution of **social_media_mins** is right-skewed. This suggests that many users spend low to moderate time on social media, while a smaller group spends significantly more time.

#### Study Minutes
The **study_mins** plot also shows a right-skewed distribution. Most users study for a moderate amount of time each day, while some users spend much longer studying.

#### Sleep Hours
The distribution of **sleep_hours** is more balanced and centered around **7 to 8 hours**. This indicates that most users have sleep durations close to a typical healthy range, with fewer extreme values.

#### Stress Level
The **stress_level** , the distribution of stress level shows a bimodal pattern, with peaks at both low and high values.it is spread across the scale but centered around a moderate level. This shows that users experience varying stress levels, though extreme stress values are less common. 

#### Digital Dependence Score
The **digital_dependence_score** distribution is slightly right-skewed, with most values concentrated in the lower to mid range. The majority of users have moderate dependence levels, while fewer users exhibit very high dependence scores.

### 2.3 Correlation Analysis

The purpose of this step is to examine the strength and direction of relationships between numerical features.

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

### Correlation Analysis Insights

The correlation heatmap reveals relationships between behavioral features and the target variable (**stress_level**) and related features.

- Device usage and phone unlocks show a strong positive correlation (0.85), indicating consistent digital behavior patterns.
- Digital dependence score is strongly correlated with device usage (0.87) and phone unlocks (0.84), confirming it as a combined indicator of digital activity.
- Device usage is negatively correlated with sleep hours (-0.59) and sleep quality (-0.66), suggesting that increased screen time reduces both sleep duration and quality.
- Sleep hours and sleep quality show a moderate positive relationship (0.41), indicating that longer sleep is associated with better sleep quality.

- Stress level shows moderate positive correlation with:
  - device usage (0.32)
  - phone unlocks (0.27)
  - notifications (0.17)
  - digital dependence score (0.35)  
  This indicates that higher digital activity is associated with increased stress.

- Stress level is negatively correlated with:
  - sleep hours (-0.29)
  - sleep quality (-0.19)  
  suggesting that better sleep helps reduce stress.

- Other features such as social media usage, study time, and physical activity show very weak correlations with stress, indicating limited direct linear relationships.

Overall, digital behavior (device usage, unlocks, notifications) and sleep patterns are the most influential factors related to stress level.

### Interpretation of Correlation Analysis

Although some features show weak correlation with the target variable, this does not necessarily mean they are unimportant.

Correlation analysis only measures **linear relationships**, while real-world data often involves **non-linear patterns and interactions between features**.

For example:
- Digital dependence is influenced by multiple combined behaviors (usage, unlocks, notifications).
- Stress may be affected by a combination of sleep, device usage, and lifestyle factors rather than a single variable.

Therefore:
- Correlation heatmaps are useful for **initial understanding** and identifying strong relationships.
- They should **not be the only method for feature selection**.
- Machine learning models can capture more complex and hidden relationships beyond linear correlations.


### 2.4 Key Insights

From the exploratory data analysis, several important insights were identified regarding digital lifestyle and stress level.

- High device usage, frequent phone unlocks, and increased notifications are associated with higher stress levels.
- Digital dependence score strongly reflects overall digital behavior and is positively related to stress.
- Sleep plays a critical role in well-being:
  - Longer sleep duration and better sleep quality are linked to lower stress levels.

- Behavioral features such as device usage, phone activity, and sleep are the most important factors influencing stress.

- In contrast, features such as social media usage, study time, and physical activity show weak individual correlations, suggesting their impact may be indirect or combined with other factors.

Overall, the analysis indicates that:

These findings support the use of machine learning models to capture complex relationships and improve prediction performance.
