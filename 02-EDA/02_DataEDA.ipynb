## 2. Explorary Data Analysis

### 2.1 Distribution Analysis
The purpose of this step is to understand how each numerical feature is distributed across the dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

cols = [
    'device_hours_per_day',
    'notifications_per_day',
    'social_media_mins',
    'study_mins',
    'sleep_hours',
    'stress_level',
    'productivity_score'
]

for col in cols:
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

### Plot Interpretation

#### Device Hours Per Day
The distribution of **device_hours_per_day** is slightly right-skewed. Most users spend around **5 to 8 hours per day** on their devices, while a smaller number of users have much higher screen time.

#### Notifications Per Day
The **notifications_per_day** feature is strongly right-skewed. Most users receive a moderate number of notifications, but a few users receive very high notification counts, creating a long tail.

#### Social Media Minutes
The distribution of **social_media_mins** is right-skewed. This suggests that many users spend low to moderate time on social media, while a smaller group spends significantly more time.

#### Study Minutes
The **study_mins** plot also shows a right-skewed distribution. Most users study for a moderate amount of time each day, while some users spend much longer studying.

#### Sleep Hours
The distribution of **sleep_hours** is more balanced and centered around **7 to 8 hours**. This indicates that most users have sleep durations close to a typical healthy range, with fewer extreme values.

#### Stress Level
The **stress_level** , the distribution of stress level shows a bimodal pattern, with peaks at both low and high values.it is spread across the scale but centered around a moderate level. This shows that users experience varying stress levels, though extreme stress values are less common. 

#### Productivity Score
The **productivity_score** distribution is relatively centered around the middle range. Most users have productivity scores near the average, with fewer very low or very high values.

### 2.3 Correlation Analysis

The purpose of this step is to examine the strength and direction of relationships between numerical features.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

### Correlation Analysis Insights

The correlation heatmap reveals relationships between behavioral features and the target variables.

- Device usage and phone unlocks show a strong positive correlation (0.85), indicating similar behavioral patterns.
- Device usage is negatively correlated with sleep hours (-0.59) and sleep quality (-0.66), suggesting that increased screen time reduces both sleep duration and quality.
- Stress level shows moderate positive correlation with device usage (0.32), phone unlocks (0.27), and notifications (0.17), indicating that higher digital activity is associated with increased stress.
- Stress level is negatively correlated with sleep hours (-0.29) and sleep quality (-0.19), suggesting that better sleep reduces stress.

In contrast, productivity score shows very weak correlations with most features, indicating that productivity is influenced by multiple factors rather than a single dominant variable.

Overall, behavioral features such as device usage and sleep are key factors affecting stress, while productivity appears to be more complex and less directly influenced by individual features.


### Interpretation of Correlation Analysis

Although some features show weak correlation with the target variables, this does not necessarily mean they are unimportant.

Correlation analysis only measures linear relationships between variables, while real-world data often involves non-linear patterns and complex interactions between features.

For example, factors such as sleep, device usage, and lifestyle may influence stress and productivity when combined, rather than individually.

Therefore, correlation heatmaps are useful for initial understanding and identifying strong relationships, but they should not be the only criterion for feature selection.

Machine learning models are capable of capturing more complex patterns beyond simple linear relationships.

### 2.5 Key Insights

From the exploratory data analysis, several important insights were identified regarding the impact of digital lifestyle on stress and productivity.

- High device usage, frequent phone unlocks, and large numbers of notifications are associated with increased stress levels.
- Sleep plays a significant role in well-being, where better sleep duration and quality are linked to lower stress levels.
- Behavioral features such as device usage and sleep are key factors influencing stress.

- In contrast, productivity score does not show strong correlation with any single feature, indicating that productivity is influenced by multiple factors rather than one dominant variable.
- Study time shows a slight positive relationship with productivity, but the effect is relatively weak.
- This suggests that productivity is more complex and depends on a combination of behaviors rather than a single factor.

Overall, the analysis confirms that digital lifestyle habits have a stronger and more direct impact on stress level, while productivity is influenced by more complex and combined factors.

These findings support the use of machine learning models to capture complex relationships between features and improve prediction performance.
