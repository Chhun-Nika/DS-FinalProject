# Library Import

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# load dataset

```python
df = pd.read_csv("../00-dataSource/dataSourceV2/clean_data_v1.csv")
```

## 2. Explorary Data Analysis

### 2.1 Distribution Analysis
The purpose of this step is to understand how each numerical feature is distributed across the dataset.

```python

# Correct numerical columns
cols = [
    'Age',
    'Daily_Screen_Time(hrs)',
    'Sleep_Quality(1-10)',
    'Stress_Level(1-10)',
    'Days_Without_Social_Media',
    'Exercise_Frequency(week)',
    'Happiness_Index(1-10)'
]

# Plot distribution
for col in cols:
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

### Plot Interpretation

#### Age
The distribution of **Age** is fairly balanced across the range of approximately **16 to 49 years old**. There is no strong skew, which suggests that the dataset includes users from different age groups in a relatively even way.

#### Daily Screen Time (hrs)
The distribution of **Daily_Screen_Time(hrs)** is approximately **bell-shaped** and centered around **5 to 7 hours per day**. Most users have moderate screen time, while only a few users have very low or very high daily screen usage.

#### Sleep Quality (1–10)
The distribution of **Sleep_Quality(1-10)** is concentrated around **5 to 7**, indicating that most users report **moderate sleep quality**. Very low and very high sleep quality scores appear less frequently.

#### Stress Level (1–10)
The distribution of **Stress_Level(1-10)** is centered mainly around **6 to 8**, showing that many users experience **moderate to high stress levels**. Very low stress scores are uncommon, while extreme high values are also less frequent.

#### Days Without Social Media
The distribution of **Days_Without_Social_Media** is slightly **right-skewed**. Most users stay away from social media for around **2 to 5 days**, while fewer users have very high numbers of days without using social media.

#### Exercise Frequency (week)
The distribution of **Exercise_Frequency(week)** is concentrated around **2 to 4 days per week**. This suggests that most users exercise occasionally or moderately, while very low and very high exercise frequencies are less common.

#### Happiness Index (1–10)
The distribution of **Happiness_Index(1-10)** is **left-skewed**, with most values clustered around **8 to 10**. This indicates that the majority of users report relatively **high happiness levels**, while low happiness scores are rare.

### 2.4 Correlation Analysis

The purpose of this step is to examine the strength and direction of relationships between numerical features.

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

### Correlation Analysis Insights

The correlation heatmap reveals relationships between behavioral features and key variables such as **stress level** and **happiness index**.

- **Daily screen time** shows a strong positive correlation with **stress level (0.74)**, indicating that higher screen usage is associated with increased stress.

- **Daily screen time** is strongly negatively correlated with:
  - **sleep quality (-0.76)**
  - **happiness index (-0.71)**  
  This suggests that excessive screen time may reduce both sleep quality and overall happiness.

- **Sleep quality** has:
  - a strong negative correlation with **stress level (-0.58)**  
  - a strong positive correlation with **happiness index (0.68)**  
  This indicates that better sleep is associated with lower stress and higher happiness.

- **Stress level** shows a strong negative correlation with **happiness index (-0.74)**, meaning that as stress increases, happiness significantly decreases.

- Features such as:
  - **age**
  - **days without social media**
  - **exercise frequency**  
  show **very weak correlations** with other variables, indicating limited direct linear relationships.


### Interpretation of Correlation Analysis

Although some features show weak correlation with the target variable, this does not necessarily mean they are unimportant.

Correlation analysis only measures **linear relationships**, while real-world behavior often involves **non-linear patterns and interactions**.

For example:
- Stress is influenced by a combination of **screen time, sleep quality, and lifestyle habits**
- Happiness is affected by both **mental state (stress)** and **physical factors (sleep quality)**

Therefore:
- Correlation heatmaps are useful for **initial understanding** and identifying strong relationships  
- They should **not be the only method for feature selection**  
- Machine learning models can capture **more complex relationships** beyond linear correlations
